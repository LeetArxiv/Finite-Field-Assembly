```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
data = np.loadtxt("swissroll_2d5000.txt")
X, Z = data[:, 0], data[:, 1]  # Extract X and Z

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(X, Z, c=np.arctan2(Z, X), cmap="rainbow", s=5, alpha=0.7)
plt.colorbar(label="Angle")
plt.title("2D Swiss Roll Visualization")
plt.xlabel("X")
plt.ylabel("Z")
plt.axis("equal")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# File names
files = ["swissroll_2d100.txt", "swissroll_2d500.txt", "swissroll_2d5000.txt"]

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, file in enumerate(files):
    data = np.loadtxt(file)
    X, Z = data[:, 0], data[:, 1]
    angle = np.arctan2(Z, X)

    sc = axes[i].scatter(X, Z, c=angle, cmap="rainbow", s=5, alpha=0.7)
    axes[i].set_title(f"Swiss Roll {i + 1}")
    axes[i].set_xlabel("X")
    axes[i].set_ylabel("Z")
    axes[i].axis("equal")

plt.tight_layout()
plt.show()
```
